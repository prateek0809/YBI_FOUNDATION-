# **Problem 4. Count number of female survived and not survived as per passenger class**

## `female[[---]].value_counts()`
## `female.groupby(--).count()`

image.png

# **Solution 4.**

```python
female[female['survived'] == 0].groupby('pclass').count()
```

```python
female[female['survived'] == 1].groupby('pclass').count()
```

# **Problem 5. Select passengers with age > 18 years**

## `titanic.`

# **Solution 5.**

```python
titanic.loc[titanic['age']>18, :]
```

# **Problem 6. Select passengers with age > 18 years and female**

## `titanic.loc[---]`

image.png

# **Solution 6.**

```python
titanic[(titanic['age'] > 18) & (titanic['sex'] == 'female')]
```

image.png

# **Problem 7. Select passengers with age > 18 years and female and travelling by first class**

## `titanic.loc[---]`

image.png

# **Solution 7.**

```python
titanic[(titanic['age'] > 18) & (titanic['sex'] == 'female') & (titanic['pclass'] == 1)]
```

image.png

# **Problem 8. Select passengers with age > 18 years and female and travelling by first class and calculate their mean age**

## `titanic.loc[---].mean()`

image.png

# **Solution 8.**

```python
titanic.loc[( titanic['age'] > 18 ) & ( titanic['sex'] == 'female') & ( titanic['pclass'] == 1), 'age'].mean()
```

image.png

# **Problem 9. Select passengers with age > 18 years or female and travelling by first class**

## `titanic.loc[---]`

image.png

# **Solution 9.**

```python
titanic.loc[( titanic['age'] > 18) | (titanic['sex'] == 'female' ) & ( titanic['pclass'] == 1), :]
```

image.png

# **Problem 10. Select passengers with age > 18 years or female and travelling by first class and calculate their mean age**

## `titanic.loc[---].mean()`

image.png

# **Solution 10.**

```python
titanic.loc[(titanic['age'] > 18 ) | ( titanic['sex'] == 'female' ) & ( titanic['pclass'] == 1), 'age'].mean()
```
